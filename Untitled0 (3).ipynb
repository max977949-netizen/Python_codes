```python
# correrlo en un cuaderno de google colab
# =====================================================================
# 🎤 CELDA 3: GEMANIS 0.01 (Conexión a GitHub + Extractor de Links Web)
# =====================================================================
import random
import requests
import json
import base64
from bs4 import BeautifulSoup  # <-- Librería para leer páginas web
from google.colab import userdata


# ---------------------------------------------------------------------
# 🔑 CONFIGURACIÓN DE SEGURIDAD
# ---------------------------------------------------------------------
GITHUB_TOKEN = userdata.get('GITHUB_TOKEN')
USUARIO_GITHUB = "max977949-netizen"
REPOSITORIO = "Python"
RUTA_ARCHIVO = "Saludos.json"

url_api = f"https://api.github.com/repos/{USUARIO_GITHUB}/{REPOSITORIO}/contents/{RUTA_ARCHIVO}"

headers = {
    "Authorization": f"token {GITHUB_TOKEN}",
    "Accept": "application/vnd.github.v3+json"
}

# 🌐 FUNCIÓN PARA EXTRAER TEXTO DE UN ENLACE (El extractor de Gemanis)
def extraer_informacion_web(url):
    try:
        print(f"🌐 Gemanis está navegando en: {url}...")
        # Hacemos la petición a la web simulando un navegador común
        headers_web = {'User-Agent': 'GemanisBot/0.01(max977949-netizen; proyecto escolar de IA)'}
        respuesta_web = requests.get(url, headers=headers_web, timeout=5)

        if respuesta_web.status_code == 200:
            soup = BeautifulSoup(respuesta_web.text, 'html.parser')
            # Extraemos los párrafos de texto de la página
            parrafos = soup.find_all('p')
            texto_completo = [p.get_text() for p in parrafos if len(p.get_text()) > 30]

            if texto_completo:
                # Tomamos los primeros 2 párrafos más importantes para no saturar
                resumen = "\n\n".join(texto_completo[:2])
                return resumen
            else:
                return "Encontré la página, pero no pude extraer texto legible."
        else:
            return f"No pude abrir el enlace. Código de error web: {respuesta_web.status_code}"
    except Exception as e:
        return f"Error al intentar leer la página web: {e}"

# 1. 📡 DESCARGAR BASE DE DATOS DESDE GITHUB
print("📡 Conectando con el cerebro de Gemanis en GitHub...")
base_de_datos = {}
file_sha = ""

try:
    r = requests.get(url_api, headers=headers)
    if r.status_code == 200:
        datos_github = r.json()
        file_sha = datos_github['sha']
        contenido_texto = base64.b64decode(datos_github['content']).decode('utf-8')
        base_de_datos = json.loads(contenido_texto)
        print(f"✨ ¡Éxito! Se cargaron {len(base_de_datos)} respuestas desde GitHub.")
    else:
        print(f"⚠️ GitHub respondió con código: {r.status_code}. Usando base de datos vacía.")
except Exception as e:
    print(f"⚠️ Error de conexión: {e}.")

print("=" * 60)

# 2. Diccionario de respuestas emocionales (Neuronal)
respuestas_emocionales = {
    0: ["🤖 Gemanis dice (Neuronal): Lamento que te sientas así. ¿Quieres que programemos algo?", "🤖 Gemanis dice (Neuronal): ¡Ánimo! Sé que vas a salir de esta."],
    1: ["🤖 Gemanis dice (Neuronal): ¡Qué excelente vibra traes hoy!", "🤖 Gemanis dice (Neuronal): ¡Esooo! Da gusto ver que estás feliz."],
    2: ["🤖 Gemanis dice (Neuronal): Deberías descansar un rato y recuperarte. ¡La salud es primero!"]
}

# 3. ENTRADA DEL USUARIO
entrada_usuario = input("\nEscribe qué quieres saber o cómo te sientes: ")
frase_limpia = entrada_usuario.lower().replace("?", "").replace("¿", "").replace("!", "").replace("¡", "").strip()

# 4. 🔍 BÚSQUEDA INTEGRADORA DE GEMANIS 0.01
if frase_limpia in base_de_datos:
    print("\n" + "=" * 60)
    print(f"📝 Frase analizada: '{entrada_usuario}'")

    contenido_guardado = base_de_datos[frase_limpia]

    # ¿Lo que guardamos en GitHub es un link de internet?
    if contenido_guardado.startswith("http://") or contenido_guardado.startswith("https://"):
        print(f"🔍 Origen: Link en GitHub 📡 -> Analizando web en tiempo real...")
        print("-" * 60)
        # Gemanis activa su extractor de páginas web
        info_extraida = extraer_informacion_web(contenido_guardado)
        print(f"\n🤖 Gemanis analizó el sitio y encontró esto:\n\n{info_extraida}")
    else:
        # Si es texto normal guardado en GitHub
        print(f"🔍 Origen: Base de Datos de GitHub (Texto Plano) 💾")
        print("-" * 60)
        print(f"🤖 Gemanis dice: {contenido_guardado}")

    print("=" * 60)

else:
    # ✍️ APRENDIZAJE AUTOMÁTICO (Puedes guardar texto O un link de Wikipedia/Web)
    print("\n" + "=" * 60)
    print(f"🤔 Gemanis dice: No tengo información sobre '{frase_limpia}'.")
    print("¿Quieres enseñarme qué responder o darme un link de internet para que yo lo investigue?")
    opcion = input("Escribe 'S' para enseñarme, o presiona ENTER para usar mi red neuronal: ").strip().lower()

    if opcion == 's':
        nueva_respuesta = input(f"Escribe una respuesta de texto O pega un link (ej. de Wikipedia) sobre '{frase_limpia}': ").strip()

        # Guardamos en nuestra memoria
        base_de_datos[frase_limpia] = nueva_respuesta

        # Codificación para GitHub
        nuevo_contenido_json = json.dumps(base_de_datos, indent=4, ensure_ascii=False)
        contenido_base64 = base64.b64encode(nuevo_contenido_json.encode('utf-8')).decode('utf-8')

        payload = {
            "message": f"Gemanis aprendió: '{frase_limpia}'",
            "content": contenido_base64,
            "sha": file_sha
        }

        print("📡 Subiendo nueva memoria a GitHub...")
        r_put = requests.put(url_api, headers=headers, json=payload)

        if r_put.status_code in [200, 201]:
            print("✨ ¡LOGRADO! He guardado esta información de forma permanente en tu GitHub.")
            if nueva_respuesta.startswith("http"):
                print(f"🤖 Gemanis dice: He memorizado que debo buscar información en este link: {nueva_respuesta}")
            else:
                print(f"🤖 Gemanis dice: He memorizado la respuesta: {nueva_respuesta}")
        else:
            print(f"⚠️ No pude guardar en GitHub. Error: {r_put.status_code}")
        print("=" * 60)

    else:
        # 🧠 Red Neuronal
        # (Aquí usamos tu modelo de TensorFlow si tienes cargadas las variables 'tokenizer', 'model', 'frases_listas')
        try:
            from tensorflow.keras.preprocessing.sequence import pad_sequences
            import numpy as np

            secuencia_prueba = tokenizer.texts_to_sequences([entrada_usuario])
            prueba_lista = pad_sequences(secuencia_prueba, maxlen=frases_listas.shape[1], padding='post')

            predicciones = model.predict(prueba_lista, verbose=0)[0]
            clase_ganadora = np.argmax(predicciones)
            probabilidad_ganadora = predicciones[clase_ganadora]

            respuesta_elegida = random.choice(respuestas_emocionales[clase_ganadora])

            print(f"🧠 Origen: Red Neuronal (Confianza: {probabilidad_ganadora:.2%}) ⚡")
            print("-" * 60)
            print(respuesta_elegida)
        except NameError:
            # Por si corres la celda sin haber entrenado el modelo antes en esa sesión
            print("🧠 Origen: Red Neuronal fuera de servicio (entrena el modelo en la Celda 1 primero).")
        print("=" * 60)
```
